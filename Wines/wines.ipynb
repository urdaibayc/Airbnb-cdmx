```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = pd.read_csv('winemag-data-130k-v2.csv', index_col=[0])
```

```python
w.country.unique()
```

```python
test_country = w.loc[w.country == 'Spain']
```

```python
test_country_2 = w.loc[w.country == 'US']
```

```python
plt.scatter(test_country.points,test_country.price)
plt.title('Spain')
plt.show()
```

```python
w.points.value_counts().head(10).plot(kind='bar')
```

```python
w.country.value_counts().head(10).plot(kind='bar')
```

```python
# create data
x = np.random.rand(40)
y = np.random.rand(40)
z = np.random.rand(40)
 
# use the scatter function
plt.scatter(x, y, s=z*1000, alpha=0.5)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(test_country.points,test_country.price)
plt.title('Spain')

plt.subplot(1,2,2)
plt.scatter(test_country_2.points,test_country_2.price, color='green' )
plt.title('US')

plt.suptitle('sub-plots')
plt.show()
```
